### **Creo una copia di lavoro**

Per evitare modifiche dirette al dataset originale, viene creata una copia del dataframe. In questo modo, tutte le operazioni di pulizia e trasformazione vengono effettuate su una versione di lavoro separata, preservando i dati raw iniziali.

```python
hr_clean = hr.copy()
```

### **Conversione in Datetime ed Estrazione Anno/Mese**

La colonna Review_Date viene convertita in formato datetime, così da rendere possibile l’estrazione di informazioni temporali in modo corretto. Questa operazione è necessaria per costruire variabili derivate utili alle analisi sull’andamento delle recensioni nel tempo.

A partire dalla data della recensione vengono estratte le variabili Review_Year e Review_Month. Queste nuove colonne consentono di effettuare analisi temporali più dettagliate, sia in SQL sia in pandas, come il conteggio delle recensioni per mese e per anno.  

```python
hr_clean['Review_Date']  = p.to_datetime(hr_clean['Review_Date'])
hr_clean['Review_Year']  = hr_clean['Review_Date'].dt.year
hr_clean['Review_Month'] = hr_clean['Review_Date'].dt.month

hr_clean.dtypes 
```

### **Rimozione duplicati esatti**

Una volta identificati, i duplicati esatti vengono rimossi dal dataframe pulito. Questo garantisce che ciascuna riga rappresenti una recensione unica e migliora l’affidabilità delle elaborazioni successive.  

```python
hr_clean.drop_duplicates(inplace=True)
```

```python
hr_clean.dropna(inplace=True)  

hr_clean.isnull().sum()
```

### **Verifica che siano stati rimossi i duplicati**

```python
hr_clean.duplicated().sum()
```

 Al termine di questo passaggio, la base dati risulta pronta per la creazione di nuove variabili e per il successivo caricamento nel database SQL.

### **Creazione delle variabili derivate**

###  Creazione della prima variabile derivata: tot recensioni per durata del soggiorno

Viene quindi creata la variabile stay_nights, ricavata dal campo Tags, per rappresentare il numero di notti di soggiorno dichiarato dal recensore. Questa nuova informazione arricchisce il dataset e può essere utile per analizzare il comportamento degli utenti in relazione ai punteggi assegnati.

```python
hr_clean['stay_nights'] = hr_clean['Tags'].str.extract(r'Stayed\s+(\d+)\s+nights?')[0]

hr_clean['stay_nights'] = p.to_numeric(hr_clean['stay_nights'], errors='coerce').astype('Int64')

print(hr_clean['stay_nights'].value_counts().sort_index())
```

### Creazione della seconda variabile derivata: Lunghezza totale della recensione

Successivamente viene costruita la variabile total_review_length, ottenuta come somma della lunghezza della recensione positiva e della recensione negativa. Questo indicatore misura l’estensione complessiva del testo e sarà utile per le analisi dedicate alla relazione tra lunghezza della recensione e score.

```python
hr_clean['total_review_length'] = (

    hr_clean['Positive_Review'].str.len() +

    hr_clean['Negative_Review'].str.len()

)

print(hr_clean['total_review_length'])   
```
